```python
import pandas as pd
import glob
```

```python
path = 'res_gossiping'
data = []
for articlePath in glob.glob(path + '/*txt'):
    with open (articlePath ,'r',encoding='utf-8') as f:
        data.append(f.read().split('---split---')[1].split('\n')[1:-1])
data
```

```python
columns =[c.split(': ')[0] for c in data[0]]
columns
```

```python
df = pd.DataFrame(data=data,columns=columns)
df
```

```python
def columnFilter(s):
    output = s.split(': ')[1]
    return output
```

```python
df['推'].apply(columnFilter)  #試試看:把"推"這一欄，放進這個過濾器裡會有什麼結果
```

```python
df2 = df.copy()  #備份df ，因等等要再教另一個過濾器
df2
```

```python
for c in df:    #直接印會印出欄位而已
    print(c)
```

```python
#寫法1
for c in df:
    df[c]= df[c].apply(columnFilter)
df
```

```python
#寫法2
for c in df2:
    df2[c] = df2[c].apply(lambda x: x.split(': ')[1])
df2
```

```python
df.to_csv('ptt.csv',index =False)
```
