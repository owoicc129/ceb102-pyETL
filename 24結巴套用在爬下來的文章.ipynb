```python
import jieba
import os
import glob
```

```python
path = 'res_gossiping'
os.listdir(path) #列出所有目錄裡的內容檔案
```

```python
glob.glob(path + '/*txt') #列出這路徑下的所有txt檔
```

```python
sb ='' #string builder
for articlePath in glob.glob(path + '/*txt'):
    with open (articlePath ,'r',encoding='utf-8') as f:
        sb += f.read().split('---split---')[0]   #把所有文章內容都倒出來
sb
```

```python
sbCut = jieba.cut(sb) #把所有文字拿去切
```

```python
wordList = [w for w in sbCut] #切完之後一個一個丟出來，並且讓他成為list
```

```python
for i in sbCut:
    print(i)    #發現用print的出來不是list 所以不好
```

```python
wordList
```

```python
wordCountDict = dict()
for w in wordList:
    if w in wordCountDict:
        wordCountDict[w] +=1  #重複出現 就+1
    else:
        wordCountDict[w] =1     #第一次出現時 一定都不在字典裡，所以讓他建立key，並且value=1
wordCountDict       
```

```python
wordCountList = [(k,v) for k,v in wordCountDict.items()]  #印出這個字典的配對，而且會是list型式
wordCountList
```

```python
wordCountList.sort(key = lambda x: x[1],reverse=True) #由大到小排列  
#隨便給一個變數叫key ,x=('作者', 348)，取x的第一個索引
```
